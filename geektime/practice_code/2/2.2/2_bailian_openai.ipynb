```python
import os
from openai import OpenAI

client = OpenAI(
  api_key = os.getenv("DASHSCOPE_API_KEY"),
  base_url = "https://dashscope.aliyuncs.com/compatible-mode/v1",
)

def get_response(prompt):
  """
    获取AI助手的回复

    Args:
        prompt (str): 用户输入的问题

    Returns:
        str: AI助手的回复内容
  """
  try:
    completion = client.chat.completions.create(
      model = "qwen-plus",
      messages = [
        {'role': 'system', 'content': '你是一个AI助手，专门回答新员工关心的公司规章制度，你的名字叫做贾维斯，请你用中文回答。'},
        {'role': 'user', 'content': prompt}
      ]
    )
    return completion.choices[0].message.content
  except Exception as e:
    return f"Error Message: {e}\n please read reference: https://help.aliyun.com/zh/model-studio/developer-reference/error-code"

# 使用示例:
response = get_response("公司的年假是怎么规定的？")
print(response)

# 即使问题相同，每次的回答也会不一样，因为模型是基于概率生成的
# 模型的用途、名称、消息角色都可以自定义
# 以上是批量生成，以下是流式生成
# 流式生成 SSE？
```

```python
import os
import time
from openai import OpenAI
from datetime import datetime

client = OpenAI(
  api_key = os.getenv("DASHSCOPE_API_KEY"),
  base_url = "https://dashscope.aliyuncs.com/compatible-mode/v1"
)

"""批量生成函数"""
def batch_generate(client, prompt):
  start_time = datetime.now()

  try:
    completion = client.chat.completions.create(
      model = "qwen-plus",
      messages = [
        {"role": "system", "content": "You are a helpful assistant." },
        {"role": "user", "content": prompt}
      ]
    )
    print("\nbatch generate result:")
    print(completion.choices[0].message.content)
  except Exception as e:
    print(f"batch generate Error: {e}")

  end_time = datetime.now()
  print(f"\nbatch generate spend: {(end_time - start_time).total_seconds():.2f} seconds")

"""流式生成函数"""
def stream_generate(client, prompt):
  start_time = datetime.now()

  try:
    stream = client.chat.completions.create(
      model = "qwen-plus",
      messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt}
      ],
      stream=True
    )
    print("\nstream generate result:")
    for chunk in stream:
      if chunk.choices[0].delta.content is not None:
        print(chunk.choices[0].delta.content, end="")
  except Exception as e:
    print(f"stream generate error: {e}")

  end_time = datetime.now()
  print(f"\nstream generate spend: {(end_time - start_time).total_seconds():.2f} seconds")

# 测试代码
prompt = "请用300字描述人工智能的发展历史"

print("Start Comparing Test ... \n")

# 同时执行两种生成方式
from threading import Thread

thread1 = Thread(target=batch_generate, args=(client, prompt))
thread2 = Thread(target=stream_generate, args=(client, prompt))

thread1.start()
thread2.start()

thread1.join()
thread2.join()
```
